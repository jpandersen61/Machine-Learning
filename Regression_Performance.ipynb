# Assignment 4

````python
def Sum_Of_Squares(x, y, hFunc):
    result = 0
    for numX, numY in zip(x,y):
        dif = hFunc(numX) - numY
        result = result + dif**2
    return result
````

````python
Sum_Of_Squares(x, y, h)
````

# Assigment 5

````python
def RMSE(x, y, hFunc):
    result=0
    MSE = 0
    n=len(x)
    if n > 0:
      MSE= Sum_Of_Squares(x, y, hFunc)/n

    return math.sqrt(MSE)

````
